```python
import pandas as pd
import numpy as np
data = pd.read_csv('complaints.csv')
data['Day'] = pd.to_datetime(data['Day'])
data.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

fig, axs = plt.subplots(ncols=3, figsize=(30,5))
sns.pointplot(y="pending", x="Day", data=data, color="red", ax=axs[0])
sns.pointplot(y="inprogress", x="Day", data=data, color="blue", ax=axs[1])
sns.pointplot(y="complete", x="Day", data=data, color="green", ax=axs[2])

```

```python
data['net_pending']= data['pending'] - data['pending'].shift()
data['net_progress']= data['inprogress'] - data['inprogress'].shift()
data['net_complete']= data['complete'] - data['complete'].shift()
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(30,5))
sns.pointplot(y="net_pending", x="Day", data=data, color="red", ax=axs[0])
sns.pointplot(y="net_progress", x="Day", data=data, color="blue", ax=axs[1])
sns.pointplot(y="net_complete", x="Day", data=data, color="green", ax=axs[2])
```
